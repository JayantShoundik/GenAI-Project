```python
os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')

```

```python
model_list = [
    "gemini/gemini-3-flash-preview", # Priority 1 (Smartest/Newest)
    "gemini/gemini-2.5-flash",       # Priority 2 (Very Stable/Reliable)
    "gemini/gemini-2.0-flash"        # Priority 3 (The Workhorse)
]
```

```python
messages = [
    {"role": "system", "content": "You are a top-class Indian stock trader. Focus on Bank Nifty."},
    {"role": "user", "content": "Explain a simple 5-minute scalp setup for today's market."}
]
```

```python
try:
    print(f"Attempting to reach our Trading Mentor...")
    response = litellm.completion(
        model=model_list[0],
        messages=messages,
        fallbacks=model_list[1:] # If model[0] fails, it tries others automatically
    )

    print(f"\n[Success using Model: {response.model}]")
    print("-" * 30)
    print(response.choices[0].message.content)

except Exception as e:
    print(f"All models failed. Market is too volatile! Error: {e}")
```
